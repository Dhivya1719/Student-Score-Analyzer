```python
#importing the pandas library

import pandas as pd

#reading the CSV file into a DataFrame
df=pd.read_csv('student-scores.csv')

#Displaying the first few rows of the DataFrame
df.head()
```

```python
#adding the total score column

df['Total Score']=df['math_score']+df['history_score']+df['physics_score']+df['chemistry_score']+df['biology_score']+df['english_score']+df['geography_score']
df.head()
```

```python
#displaying the DataFrame information and columns

df.info()
df.columns
```

```python
# highest marks per subject

subject_columns=['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']
high=df[subject_columns].max()
print("Highest marks per subject:\n",high)
```

```python
#lowest marks per subject

low=df[subject_columns].min()
print("Lowest marks per subject:\n",low)
```

```python
#display with names

df['Full Name'] = df['first_name'] + ' ' + df['last_name']
print(df[['Full Name', 'Total Score']].head())
```

```python
import matplotlib.pyplot as plt

avg_marks = df[subject_columns].mean()
avg_marks.plot(kind='bar', color='skyblue')
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_10 = df.sort_values(by='Total Score', ascending=False).head(10)
print(top_10[['Full Name', 'Total Score']])
```
